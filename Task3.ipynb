# **4. Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
num_cols = ['Elevation','Aspect','Slope',
            'Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology',
            'Horizontal_Distance_To_Roadways','Hillshade_9am','Hillshade_Noon',
            'Hillshade_3pm','Horizontal_Distance_To_Fire_Points']

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

df[num_cols].head()
```

# **5. spliting data**

```python
#first we drop the output coloumn from x and let it be only in y

x = df.drop('Cover_Type', axis=1)
y = df['Cover_Type']
```

```python
#spliting
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# **Modeling**

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
xgb_model = XGBClassifier()

# Adjust target variable to be 0-indexed
y_train_adjusted = y_train - 1
y_test_adjusted = y_test - 1

xgb_model.fit(x_train, y_train_adjusted)
```

```python
from xgboost import XGBClassifier, plot_importance



# استخدم الموديل نفسه
plot_importance(xgb_model.get_booster())
plt.show()
```

```python
y_pred = xgb_model.predict(x_test)
y_pred[:10]
```

```python
y_test_adjusted[:10]
```

##**Accuracy scores**

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
acc = accuracy_score(y_test_adjusted, y_pred)
train_acc = accuracy_score(y_train_adjusted, xgb_model.predict(x_train))
print(f"XGBoost Accuracy: {acc:.4f}")
print(f"XGBoost Train Accuracy: {train_acc:.4f}")
```

```python
print("\nClassification Report:")
print(classification_report(y_test_adjusted, y_pred))
```

```python
# Confusion Matrix Visualization
cm = confusion_matrix(y_test_adjusted, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("XGBoost Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate all Precision socres, ,  for each class
precision_per_class = precision_score(y_test_adjusted, y_pred, average=None)



print("Precision per class:", precision_per_class)
print("Precision (Macro):", precision_score(y_test_adjusted, y_pred, average='macro'))
print("Precision (Weighted):", precision_score(y_test_adjusted, y_pred, average='weighted'))


```

```python
# Calculate all Recall socres
recall_per_class = recall_score(y_test_adjusted, y_pred, average=None)
print("Recall per class:", recall_per_class)
print("Recall (Macro):", recall_score(y_test_adjusted, y_pred, average='macro'))
print("Recall (Weighted):", recall_score(y_test_adjusted, y_pred, average='weighted'))
```

```python
# Calculate all F1-socre socres
f1_per_class = f1_score(y_test_adjusted, y_pred, average=None)
print("F1-score per class:", f1_per_class)
print("F1-score (Macro):", f1_score(y_test_adjusted, y_pred, average='macro'))
print("F1-score (Weighted):", f1_score(y_test_adjusted, y_pred, average='weighted'))
```

## hyperparapeter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#  Hyperparameter Tuning for Random forest
rf_params = {
    'n_estimators': [50, 100, 150],  # قللنا عدد الـ trees
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}
```

```python
rf_random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=rf_params,
    n_iter=3,      # كان 5 --> قللناه لـ 3
    cv=2,          # كان 3 --> قللناه لـ 2
    verbose=1,
    random_state=0,
    n_jobs=-1
)

# Train
rf_random_search.fit(x_train, y_train_adjusted)

```

```python

best_rf = rf_random_search.best_estimator_
print("Best Parameters:", rf_random_search.best_params_)
```

```python
print("Best Random Forest Parameters:", rf_random_search.best_params_)
print(f"Best Random Forest CV Score: {rf_random_search.best_score_:.4f}")
```

```python
best_rf_pred = best_rf.predict(x_test)
print(f"Tuned RF Test Accuracy: {accuracy_score(y_test_adjusted, best_rf_pred):.4f}")
```
